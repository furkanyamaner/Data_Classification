```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option("display.max_rows", None)

dFrame = pd.read_excel("Data_processed.xlsx",)

def cleanData(dataFrame):
    
    dataFrame.replace(0.00000, np.nan, inplace=True,)
    
    
    for col in dataFrame.columns:
        if dataFrame[col].isnull().any():  
            if dataFrame[col].dtype in ['float64', 'int64']: 
                dataFrame[col].fillna(dataFrame[col].mean(), inplace=True)
            else:  
                dataFrame[col].fillna(dataFrame[col].mode()[0], inplace=True)

        print("Veri temizleme işlemi tamamlandı.")
        print(dataFrame[col].head(10))



cleanData(dFrame)

"""

Bu kısımda tüm veriler küsüratlı olarak çıkıyor.
Küsüratlı veriler F1 skorunu olumlu yönde etkiler.
İkisinden birini seçmeliyiz.


nullValue = 0.000000

if nullValue in dataFrame[col].values:
    dataFrame[col] = dataFrame[col].mean()



"""

"""Bu kısımda sadece 0.00000 olanların yerini ilgili sütunun mean'i ile dolduruyorum.
NaN veya null olan yerler varsa onlar da kontrol edilmelidir."""
```
